```python
import pandas as pd
#nhập data
df={
    "Name": ['Mr. Harris', 'Mrs. Bradley', 'Miss. Laina', 'Mrs. Health', 'Mr. William', 'Mr. James', 'Mr. Timothy', 'Mrs. Oscar', 'Mrs. Nicholas', 'Miss. Marguerite', 'Mr. William', 'Miss. Hulda', 'Mr. Lawrence', 'Mr. Wilam'],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.9, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "AGE": [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}
# tạo data frame
df = pd.DataFrame(df)
df
```

```python
#3. tính giá trị trung bình cột "AGE" lấy số nguyên
trungbinh_AGE = df["AGE"].mean()
trungbinh_AGE
```

```python
#4. nhóm dữ liệu cột "Height" và tính tổng trung bình cột "Weight"

HeightWeight= df.groupby('Height')['Weight'].mean()
HeightWeight
```

```python
#5. sắp xếp dataframe theo cột "Weight" tăng dần 

Weight_tangdan = df.sort_values(by='Weight', ascending=True)
Weight_tangdan
```

```python
#6. vẽ biểu đồ cột cho "AGE"
import matplotlib.pyplot as plt
df['AGE'].plot(kind='bar')
plt.show()
```

```python
#7. vẽ biểu đồ phân tán cho "AGE" và "Height"

import seaborn as sns
sns.scatterplot(x='AGE', y='Height', data=df)
plt.show()
```

```python
#11. kiểm tra xem có giá trị NaN nào trong df không

df.isnull().sum()
```

```python
#12. Tìm các giá trị trùng lặp trong df dựa trên cột "Name" và loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên

df_dup_check = df.drop_duplicates(subset='Name', keep='first')
df_dup_check
```

```python
#13. thay thế các giá trị cột "AGE" lớn hơn 40 bằng giá trị trung bình của cột đó
#trungbinh_AGE ở câu 4
df.loc[df['AGE'] > 40, 'AGE'] = trungbinh_AGE
df
```

```python
#14. tạo cột mới trong df với tên "cs_BMI", tính số BMI mỗi người

df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)
df
```

```python
#15. tạo cột mới với tên "Phan_loai" với các mức "0", "1", "2", "3", tương ứng với mức chỉ số BMI trên
# chỉ số BMI dưới 18,5 là thiếu cân, 18,5 - 22.9 là bình thường, 23 - 24.9 là thừa cân, 25 trở lên là béo phì

df['Phan_loai'] = pd.cut(df['cs_BMI'], bins=[0, 18.5, 22.9, 24.9, 100], labels=['0', '1', '2', '3'])
df
```

```python
#16. chuẩn hóa giá trị của cột "cs_BMI" chuẩn hóa giá trị trong đoạn từ 0 tới 1
df
```

```python
#17. "Mr" tương ứng với "Male", "Mrs" và "Miss" tương ứng với "Female"
# tạo cột với tên "GT", sau khi thực hiện, xóa các từ "Mrs", "Mr", "Miss" nếu trong cột "Name" còn các từ đó

df['GT'] = df['Name'].str.extract('(Mr\.|Mrs\.|Miss\.)')
df['GT'] = df['GT'].replace(['Mr.', 'Mrs.', 'Miss.'], ['Male', 'Female', 'Female'])
df

```

```python
#18 vẽ biểu đồ tròn cho cột "GT"

df['GT'].value_counts().plot(kind='pie')
plt.show()
```

```python
#19 Trực quan hóa dữ liệu người nữ trên "Phan_Loai"

df_female = df[df['GT'] == 'Female']
sns.countplot(x='Phan_loai', data=df_female)
plt.show()
```
